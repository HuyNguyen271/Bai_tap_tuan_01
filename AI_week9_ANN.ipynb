```python
# 1 pixel la 1 in put
# 10 out put tuong duong 10 so
from keras.models import Sequential # ANN su ke tiep (cac lop ke tuc nha), dung cho mang ANN
from keras.layers import Dense # full_conected (ket noi deu het voi cac no ron )
model = Sequential() # dung cho mag
model.add(Dense(512, activation='relu',input_shape=(28*28,))) # them lop  (512lop)
model.add(Dense(10,activation='softmax'))

model.summary()

```

```python
train_images = train_images.reshape((60000,28*28)) #chuyen het du lieu 60000
train_images = train_images.astype('float32')/255 #chuyen dang 255 ve tu 0->1

test_images = test_images.reshape((10000,28*28))
test_images = test_images.astype('float32')/255 



# digit = train_images[50] # thay so trong 60000 of train


# plt.imshow(digit)
# plt.show()
# print("So nay la: ",train_labels[50])
```

```python
train_images.shape
```

```python
test_images.shape
```

```python
from keras.utils import to_categorical # chuyen 1 bit ra nhieu bit

train_labels = to_categorical(train_labels,10)
test_labels =to_categorical(test_labels,10)



train_labels.shape
```

```python
test_labels.shape
```

```python
from keras import layers
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, epochs=20, batch_size=128)
```

```python
test_loss,test_acc =model.evaluate(test_images,test_labels)
print(' do mat mat: ', test_loss)
print(' do chinh xac ', test_acc)
```
