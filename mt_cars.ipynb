```python
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

import pandas as pd

url = '/content/sample_data/mtcars.csv'

data = pd.read_csv(url)

model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'],data['wt']**2,data['wt']**3,data['wt']**4,data['wt']**5,data['wt']**6,data['wt']**7,data['wt']**8,data['wt']**9]).T
model.fit(X=pd.DataFrame([data['wt'],data['wt']**2,data['wt']**3,data['wt']**4,data['wt']**5,data['wt']**6,data['wt']**7,data['wt']**8,data['wt']**9]).T,y=data['mpg'])

print('Gia tri Intercept bac 9 la: ',model.intercept_)
print('Gia tri Coeffient bac 9 la: ',model.coef_)
print('Gia tri Score bac 9 la: ',model.score(X,data['mpg']))

range = np.arange(1.5,5.5,0.1)

X_range = pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9]).T
y_pred = model.predict(pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9]).T)

data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='green',xlim=(0,7))
plt.plot(range,y_pred,color='blue')



```
