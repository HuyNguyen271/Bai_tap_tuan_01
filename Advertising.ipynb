```python

import pandas as pd

url = '/content/sample_data/Advertising.csv'
data = pd.read_csv(url)
data.head()
```

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['TV']),y=data['sales'])
print('Gia tri Intercept bac 1 la: ',model.intercept_)
print('Gia tri Coeffient bac 1 la: ',model.coef_)
print('Gia tri Score la:  ',model.score(pd.DataFrame(data['TV']),data['sales']))

data.plot(kind='scatter',x='TV',y='sales',figsize=(9,9),color='red')
model.score(pd.DataFrame(data['TV']),data['sales'])

y_pred = model.predict(pd.DataFrame(data['TV']))
plt.plot(data['TV'],y_pred,color='blue')


```
